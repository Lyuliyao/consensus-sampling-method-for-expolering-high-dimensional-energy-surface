```julia
using PyPlot
using NPZ
using DelimitedFiles
using LaTeXStrings
using JLD2
using MAT
using LaTeXStrings
using DelimitedFiles
using ColorSchemes
using Statistics
```

```julia
MARKERSIZE = 7
LINEWIDTH=4
LEGENDFONTSIZE=24
LABELFONTSIZE=24
TICKFONTSIZE=24
cm = get_cmap(:tab20)
colorrange = (0:19) ./ 20
```

```julia
function contourplot1(ax,file)
    levels = 0:1:60
    data = readdlm(file,skipstart=9)
    data_size = Int64(sqrt(size(data,1)))
    phi  = reshape(data[:,1],data_size,data_size)
    psi  = reshape(data[:,2],data_size,data_size)
    ene  = reshape(data[:,3],data_size,data_size)
#     a=ax.contour(phi,psi,ene.-minimum(ene),levels,colors="k")
    ax.contourf(phi,psi,ene.-minimum(ene),levels,cmap=cm,extend="max")
#     setp(ax.get_xticklabels(),visible=false)
#     setp(ax.get_yticklabels(),visible=false)
end
function contourplot3(ax,file)
    levels =0:1:60
    data = readdlm(file)
    data_size = Int64(sqrt(size(data,1)))
    phi  = reshape(data[:,1],data_size,data_size)
    psi  = reshape(data[:,2],data_size,data_size)
    ene  = reshape(data[:,3],data_size,data_size)
#     a=ax.contour(phi,psi,ene.-minimum(ene),levels,colors="k")
    ax.contourf(phi,psi,ene.-minimum(ene),levels,cmap=cm,extend="max")
#     setp(ax.get_xticklabels(),visible=false)
#     setp(ax.get_yticklabels(),visible=false)
end
function contourplot2(ax,file)
    levels = 0:1:60
    data = npzread(file)
#     a=ax.contour(data["X"],data["Y"],data["Z"].-minimum(data["Z"]),levels,colors="k")
    h = ax.contourf(data["X"],data["Y"],data["Z"].-minimum(data["Z"]),levels,cmap=cm,extend="max")
#     setp(ax.get_xticklabels(),visible=false)
#     setp(ax.get_yticklabels(),visible=false)
    return h
end
```

```julia
cm = get_cmap(:RdBu_r)
fig = figure("pyplot_subplot_touching",figsize=(6,9))
subplots_adjust(wspace=0.0,hspace=0.0,top = 0.87, bottom = 0.87-0.77, right = 0.86, left = 0.09) # Set the vertical spacing between axes
ax11 = plt.subplot(3,2,1)
ax21 = plt.subplot(3,2,2)
ax12 = plt.subplot(3,2,3)
ax22 = plt.subplot(3,2,4)
ax13 = plt.subplot(3,2,5)
ax23 = plt.subplot(3,2,6)
contourplot1(ax11,"./example_2D_13_anvil/fes_phi1_psi1.dat")
contourplot2(ax21,"./consensus31/example_2D_example1_phi1_psi1.npz")

contourplot1(ax12,"./example_2D_12_anvil/fes_phi2_phi3.dat")
contourplot2(ax22,"./consensus31/example_2D_example1_phi2_phi3.npz")

contourplot1(ax13,"./example_2D_12_anvil/fes_psi2_psi3.dat")
h = contourplot2(ax23,"./consensus31/example_2D_example1_psi2_psi3.npz")
ax11.tick_params(direction="in")
ax21.tick_params(direction="in")
ax12.tick_params(direction="in")
ax22.tick_params(direction="in")
ax13.tick_params(direction="in")
ax23.tick_params(direction="in")

ax23.set_xticks(-2:2:3)
ax13.set_xticks(-2:2:3)
setp(ax11.get_xticklabels(),visible=false)
setp(ax12.get_xticklabels(),visible=false)
setp(ax12.get_xticklabels(),visible=false)
setp(ax22.get_xticklabels(),visible=false)

ax11.set_yticks(-2:2:3)
ax12.set_yticks(-2:2:3)
ax13.set_yticks(-2:2:3)
setp(ax21.get_yticklabels(),visible=false)
setp(ax22.get_yticklabels(),visible=false)
setp(ax23.get_yticklabels(),visible=false)


ax11.tick_params(axis="y",labelsize=TICKFONTSIZE)
ax12.tick_params(axis="y",labelsize=TICKFONTSIZE)
ax13.tick_params(axis="y",labelsize=TICKFONTSIZE)


ax13.tick_params(axis="x",labelsize=TICKFONTSIZE)
ax23.tick_params(axis="x",labelsize=TICKFONTSIZE)

pos = 2.3
ax11.text(pos,pos,"(a)",fontsize=TICKFONTSIZE,ha="center")
ax21.text(pos,pos,"(b)",fontsize=TICKFONTSIZE,ha="center")
ax12.text(pos,pos,"(c)",fontsize=TICKFONTSIZE,ha="center")
ax22.text(pos,pos,"(d)",fontsize=TICKFONTSIZE,ha="center")
ax13.text(pos,pos,"(e)",fontsize=TICKFONTSIZE,ha="center")
ax23.text(pos,pos,"(f)",fontsize=TICKFONTSIZE,ha="center")

cbar_ax = fig.add_axes([0.87, 0.1, 0.02, 0.77])
plt.colorbar(h, cax=cbar_ax, ticks=0:10:60)
cbar_ax.tick_params(labelsize=TICKFONTSIZE)
# cbar_ax.set_yticklabels([0:20:100]) 
plt.savefig("../figure/ala16_2D.eps")
plt.savefig("./ala16.pdf")
```
