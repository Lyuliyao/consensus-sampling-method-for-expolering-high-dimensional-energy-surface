```python
def compute_energy(dim1,dim2,vec):
    xlist = np.linspace(-pi, pi, 300)
    ylist = np.linspace(-pi, pi, 300)
    X, Y = np.meshgrid(xlist, ylist)
    inputs = torch.zeros(X.shape[0]*X.shape[1],dim)
    for i in range(dim):
        inputs[:,i] =vec[i];
    inputs[:,dim1] = torch.tensor(X.reshape(-1));
    inputs[:,dim2] = torch.tensor(Y.reshape(-1));
    Z = potential1.bias_potential(inputs.T,1)
    return X,Y,Z.reshape(X.shape).detach().numpy() 
```

```python
value_vec =np.zeros(dim)
value_vec[:15] = -2.5
value_vec[15:] = 2.7
X,Y,Z = compute_energy(4,19,value_vec)
np.savez("example_2D_example1_phi1_psi1.npz",X=X,Y=Y,Z=Z)
X,Y,Z = compute_energy(2,3,value_vec)
np.savez("example_2D_example1_phi3_phi4.npz",X=X,Y=Y,Z=Z)
X,Y,Z = compute_energy(15,16,value_vec)
np.savez("example_2D_example1_psi1_psi2.npz",X=X,Y=Y,Z=Z)
X,Y,Z = compute_energy(2,17,value_vec)
np.savez("example_2D_example1_phi3_psi3.npz",X=X,Y=Y,Z=Z)
X,Y,Z = compute_energy(1,2,value_vec)
np.savez("example_2D_example1_phi2_phi3.npz",X=X,Y=Y,Z=Z)
X,Y,Z = compute_energy(18,19,value_vec)
np.savez("example_2D_example1_psi4_psi5.npz",X=X,Y=Y,Z=Z)
# X,Y,Z = compute_energy(14,29,value_vec)
# np.savez("example_2D_example1_phi15_psi15.npz",X=X,Y=Y,Z=Z)
```
